```python
import numpy
import pandas as pd
df= pd.read_csv('car_dataset.csv')
df.columns
df.dtypes
```

```python
print(f'Buying{df.Buying.unique()}')
print(f'Maint{df.Maint.unique()}')
print(f'clean_Doors{df.Doors.unique()}')
print(f'Person{df.Person.unique()}')
print(f'Lug_boot{df.Lug_boot.unique()}')
print(f'Safety{df.Safety.unique()}')
print(f'classes{df.classes.unique()}')
print(f'Length of Doors{len(df.Doors.unique())}')
```

```python
df['clean_Buying']=df['Buying'].apply({'vhigh':3,'high':2,'med':1,'low':0}.get)
df['clean_Maint']=df['Maint'].apply({'vhigh':3,'high':2,'med':1,'low':0}.get)
df['clean_Doors']=df['Doors'].apply({2:2,3:3,4:4,'5more':6}.get)
df['clean_classes']=df['classes'].apply({'unacc':1,'acc':2,'good':3,'vgood':4}.get)

print(df.info)
df.head()
```

```python
df['clean_Buying']=df['Buying'].apply({'vhigh':3,'high':2,'med':1,'low':0}.get)
df['clean_Maint']=df['Maint'].apply({'vhigh':3,'high':2,'med':1,'low':0}.get)
df['clean_Doors']=df['Doors'].apply({2:2,3:3,4:4,'5more':6}.get)
df['clean_Person']=df['Person'].apply({2:2,4:4,'more':6}.get)
df['clean_Safety']=df['Safety'].apply({'low':0,'med':1,'high':2,}.get)
df['clean_classes']=df['classes'].apply({'unacc':1,'acc':2,'good':3,'vgood':4}.get)
print(f'Length of clean_Buying{len(df.clean_Buying.unique())}')
print(f'Length of clean_Maint{len(df.clean_Maint.unique())}')
print(f'Length of clean_Doors{len(df.clean_Doors.unique())}')
print(f'Length of clean_Person{len(df.clean_Person.unique())}')
print(f'Length of clean_Safety{len(df.clean_Safety.unique())}')
print(f'Length of clean_classes{len(df.clean_classes.unique())}')
df=df[['clean_Buying','clean_Maint','clean_Doors','clean_Person','clean_Safety','clean_classes']]
df.head()
```

```python
import numpy
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
dataset = df.values
X= dataset[:,0:6]
Y= dataset[:,6]
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size=0.2)
# create model
knn = KNeighborsClassifier()
knn.fit(X_Train, Y_Train)
KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski', metric_params=None, n_jobs=1, n_neighbors=5, p=2, weights='uniform')

predictions = knn.predict(X_Test)

score = accuracy_score(Y_Test, predictions)
print(score)
```
